```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("../data/enriched")  # adapte si nécessaire

dossiers = [
    "ratings.parquet",
    "movies_enriched.parquet",
    "movies.parquet",
    "links.parquet"
]

for dossier in dossiers:
    path = DATA_DIR / dossier
    if not path.exists():
        print(f"⚠️  {dossier} introuvable\n")
        continue

    # Lecture de tous les part-*.parquet dans le dossier
    df = pd.read_parquet(path)

    taille_mb = round(sum(f.stat().st_size for f in path.rglob("*.parquet")) / 1e6, 1)

    print("=" * 70)
    print(f"📁  {dossier}  ({len(df):,} lignes | {df.shape[1]} colonnes | {taille_mb} MB)")
    print("=" * 70)

    info = pd.DataFrame({
        "colonne":    df.columns,
        "dtype":      [str(df[c].dtype) for c in df.columns],
        "non_null_%": [round((1 - df[c].isnull().mean()) * 100, 1) for c in df.columns],
        "exemple":    [str(df[c].dropna().iloc[0]) if df[c].notna().any() else "N/A"
                       for c in df.columns]
    })

    print(info.to_string(index=False))
    print()
```
