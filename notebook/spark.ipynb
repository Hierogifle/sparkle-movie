```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("check-spark")
    .master("local[*]")
    .config("spark.driver.memory", "4g")
    .getOrCreate()
)

df = spark.range(0, 10_000_000)
print(df.count())

spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = (
    SparkSession.builder
    .appName("sparkle-movie")
    .master("local[*]")
    .config("spark.driver.memory", "6g")
    .getOrCreate()
)

DATA_DIR = "../data/brut"  # à adapter

ratings = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{DATA_DIR}/ratings.csv")
)

ratings.printSchema()
ratings.show(5)
print("nb_lignes:", ratings.count())
print("nb_users :", ratings.select("userId").distinct().count())
print("nb_movies:", ratings.select("movieId").distinct().count())
```
